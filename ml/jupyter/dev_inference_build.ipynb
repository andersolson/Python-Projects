Script runs the saved model from `dev_train_model.ipynb` named `wildlandwater_resnet18.pkl`. This model is used to make predictions on unseen images. The result is a table containing the shapefile grid ID, the predicted label and class probabilities of the label.

```python
from fastai.vision.all import *
import pathlib
```

#### Load the learner with the trained model

D:\Projects\Github\SAGIS-Risk\1 Natural Hazards\1 Wildfire\Firebreak_Enhancement\dev\.models\wildlandwater_resnet18.pkl

```python
#resnet18model = r'D:\Projects\Github\SAGIS-Risk\1 Natural Hazards\1 Wildfire\Firebreak_Enhancement\dev\.models\wildlandwater_resnet18.pkl' 

learn = load_learner(r'D:\Projects\Github\SAGIS-Risk\1 Natural Hazards\1 Wildfire\Firebreak_Enhancement\dev\fastai_models\wildlandwater_resnet18.pkl')
```

#### Load the new image

B:\projects\FirebreakML\Firebreak_Water_Wildland\imagery\[ST]_Water_Wildland

```python
imgPath = r'B:\projects\FirebreakML\Firebreak_Water_Wildland\imagery\CO_Water_Wildland\wildland\0.jpg'

img = PILImage.create(imgPath)

# Display the image
img.show()
```

#### Make predictions on the new image

```python
pred_class, pred_idx, outputs = learn.predict(img)
```

#### Get the predicted label

```python
predicted_label = pred_class
print("Predicted label:", predicted_label)
```

#### Get the probabilities/scores for each class


```python
class_probabilities = outputs.softmax(dim=-1)
print("Class probabilities:", class_probabilities)
```
