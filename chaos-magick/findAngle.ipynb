```python
import math
```

```python
startX = 4
startY = 5
```

```python
endX = 10
endY = -1
```

```python
myradians = math.atan2(endY-startY, endX-startX)
myradians
```

```python
mydegrees = math.degrees(myradians)
mydegrees
```

```python
import matplotlib.pyplot as plt
```

```python
x = [4, 10]
y = [5,-1]
```

```python
plt.plot(x,y,'->')
plt.axis('equal')
plt.show()
```

```python
from matplotlib.markers import MarkerStyle
```

```python
m = MarkerStyle('>')
m._transform.rotate_deg(-45)
```

```python
plt.plot(x,y)
plt.axis('equal')
plt.show()
```

```python
plt.plot(x,y)
plt.scatter(x,y, s=200, marker=m, color="crimson")
plt.axis('equal')
plt.show()
```

```python
m = MarkerStyle('>')
m._transform.rotate_deg(mydegrees)


plt.plot(x,y)
plt.scatter(x,y, s=200, marker=m, color="crimson")
plt.axis('equal')
plt.show()
```
